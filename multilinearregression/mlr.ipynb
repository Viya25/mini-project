                                        Multiple Linear Regression using 50_startups.csv                                                                        

1. Problem statement
we want to predict Profit of startup using multiple factors like R&D Spend, Administartion and Marketing spend

2. Import Required Libraries

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_absolute_error,mean_squared_error, r2_score
```

3. Load Dataset

```python
df=pd.read_csv(r"C:\Users\Asus\Ethans\ml\multilinearregression\50_Startups.csv")
df
df.head()
```

4. Dataset Overview

```python
df.shape
```

5. Exploratory data analysis

```python
plt.scatter(df['R&D Spend'],df['Profit'])
plt.xlabel('R&D Spend')
plt.ylabel('Profit')
plt.title('R&D Spend VS Profit')
plt.show()
```

6. Prepare Features and Target

```python
x=df[['R&D Spend','Administration','Marketing Spend']]
y=df['Profit']
```

7. Train-Test Split

```python
x_train, x_test, y_train,y_test=train_test_split(x,y, test_size=0.2, random_state=42)
```

8. Train the model

```python
model=LinearRegression()
model.fit(x_train,y_train)
```

9. Model Coefficients

```python
model.coef_, model.intercept_
```

10. Predection

```python
y_pred=model.predict(x_test)
```

11.Model Evaluation

```python
print("MAE",mean_absolute_error(y_test,y_pred))
print("MSE",mean_squared_error(y_test,y_pred))
print("R2",r2_score(y_test,y_pred))
```

12. Final Conclusion*****This notebook demonstrate Multiple Linear Regression using a real startup dataset
