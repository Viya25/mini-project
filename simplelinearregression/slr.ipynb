```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

```python
plt.scatter(df["cgpa"],df["package"])
plt.xlabel("CGPA")
plt.ylabel("Package(in lpa)")
```

```python
#step 4: Spitting data into  Features and Target
X=df.iloc[:,0:1]#input
Y=df.iloc[:,1:2]#output
```

```python
df.iloc[:,0:1]
```

```python
X
```

```python
#split data into training(80%) and testing (20%) sets
# X=input features(s), Y=target variable 
# random_states ensures the same split every time
from sklearn.model_selection import train_test_split
X_train,X_test,Y_Train,Y_test=train_test_split(X,Y,test_size=0.2,random_state=0)
```

```python
#160 number of rows and 1 number of col
X_test.shape
X_train.shape
```

        We use LinearRegression() from Scikit-leran.
 The model will find the best-fitting line through the data by minimizing the difference btw predicted and actual values.                       

```python
#create and train Linear Regression
#lr=model object;
from sklearn.linear_model import LinearRegression
lr=LinearRegression() #creating an object/instance of LR model
lr.fit(X_train,Y_Train)  #training of X and Y. and tarining completes here. later starts predcition
```

```python
#This is where training happens. 
#The .fit() method tells the model to learn the relationship btw X_train (your inputs/features) and Y_traim(your output/target)
#Internally, the model finds the best-fit line that minimizes the error ntw the predictions and the actual values.
#After this step>>>>>>> Your model has learned the coefficients(slope) and the intercept (where the line crosses the y-axis).
#You can chcek them using: lr.coef_        lr.intercept_
```
