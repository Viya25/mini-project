```python
import pandas as pd
```

```python
dataset = pd.read_csv(r"C:\Users\Asus\Ethans\ml\svm\Copy of Social_Network_Ads.csv")
```

```python
dataset
```

```python
dataset[['Purchased']].value_counts()
```

```python
dataset = pd.get_dummies(dataset,dtype=int,drop_first=True)
```

```python
dataset
```

```python
dataset.columns
```

```python
independent = dataset[['User ID', 'Age', 'EstimatedSalary','Gender_Male']]
```

```python
dependent = dataset[['Purchased']]
```

```python
#Split train and test sets using sklearn
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test = train_test_split(independent, dependent, test_size=0.30, random_state=0)
```

```python
X_train
```

```python
#standardisation
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
X_train = sc.fit_transform(X_train) # mean and standard deviation calculate here
X_test = sc.transform(X_test)
```

```python
X_train
```

```python
#Model creation #learning process
#procedure
from sklearn.svm import SVC
# to sggregate a model in SVM using straight line when kernel ="linear"
#rbf for non linear # radial bais function
classifier = SVC(kernel="rbf", C=1000)
classifier.fit(X_train,y_train)
```

```python
classifier.intercept_
```

```python
# to knoe how many suppot vector does the model take
classifier.n_support_
```

```python
# the model is created with 34 datapoints
classifier.support_
```

```python
y_predict = classifier.predict(X_test)
```

```python
from sklearn.metrics import confusion_matrix
cm = confusion_matrix(y_test,y_predict)
```

```python
print(cm)
```

```python
from sklearn.metrics import classification_report
clf_report = classification_report(y_test,y_predict)
```
