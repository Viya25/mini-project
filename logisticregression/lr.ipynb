```python
import numpy as numpy
import matplotlib.pyplot as plt
import pandas as pd
```

```python
dataset=pd.read_csv(r"C:\Users\Asus\Ethans\ml\logisticregression\Copy of Social_Network_Ads.csv")
dataset
```

```python
#Selecting independebt (X) and dependent (y) variables
# X= features(input data)
# iloc[:,[2,3]] means:
# ":" all rows
# "[2,3]" take 2 column and 3 column only
# These are the two inputs features we will use for prediction
X=dataset.iloc[:,[2,3]].values
y=dataset.iloc[:,4].values
```

```python
# Splitting the dataset into the training set and Test set
from sklearn.model_selection import train_test_split
X_train, X_test, y_train,y_test=train_test_split(X,y, test_size=0.25, random_state=42)
```

```python
"""
Quick rule for balance in classification
A dataset is considered balanced if no class is extremely larger than the other. Normally, if one classs is more than 60%-70% of the data, the dataset becomes 
imbalanced
"""
```

```python
dataset["Purchased"].value_counts()
```

```python
# Feature scaling:
# We scale features so that all values have a similar range
# StandardScaler converts the data into a standard form:
# mean= 0 and sd = 1
# This helps the model learn better and faster

from sklearn.preprocessing import StandardScaler
sc=StandardScaler()

# Training data to fit + transform
# sc.fit_transform() -->> scaler ko training data ke hisaab se sikhaata hai
# aur fir data ko scale karta hai
X_train=sc.fit_transform(X_train)
X_test=sc.transform(X_test)  #scale test data using same rules
X_train
```

```python
# Fitting Logistic Regression to Training set
from sklearn.linear_model import LogisticRegression
classifier=LogisticRegression(random_state=0)
classifier.fit(X_train,y_train)
```

```python
# Prediting the Test set results
y_pred=classifier.predict(X_test)
y_pred
```

```python
# Making thr Confusion Matrix (correct and incorrect matrix)
from sklearn.metrics import confusion_matrix, accuracy_score, f1_score, precision_score,recall_score
cm = confusion_matrix(y_test,y_pred)
accuracy_score=accuracy_score(y_test,y_pred)

print(accuracy_score)
print(cm)
print("\n Precision")
print(precision_score(y_test,y_pred))

print("\n recall:")
print(recall_score(y_test,y_pred))

print("\n f1 score")
print(f1_score(y_test,y_pred))
```
